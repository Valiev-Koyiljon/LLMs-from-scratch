### Attention

```python
import torch
```

- B - batch_size
- T - token_size  (context_size)
- C - embedding_size

```python
torch.manual_seed(123)
B, T, C = 4, 8, 32
head_size = 16

x = torch.rand(B, T, C)
x.shape
```

```python
query = torch.nn.Linear(C, head_size, bias=False)
key = torch.nn.Linear(C, head_size, bias=False)
value = torch.nn.Linear(C, head_size, bias=False)
```

```python
q = query(x)  # shape(B, T, head_size)
k = key(x)  # shape(B, T, head_size)
v = value(x)  # shape(B, T, head_size)
```

```python
q.shape, k.shape, v.shape
```

```python
k.shape, k.transpose(-2, -1).shape
# k.T.shape 
```

```python
q # [4, 8, 16] @ [4, 16, 8] --> [4, 8, 8]
```

```python
atten_score = torch.matmul(q, k.transpose(-2, -1))
atten_score.shape

atten_weights = torch.softmax(atten_score/k.shape[-1]**0.5, dim=-1)
atten_weights[0][5].sum()
```

```python
atten_weights = torch.softmax(atten_score, dim=-1)
atten_weights[0][5]
```

```python
atten_score[0, 5]
```

```python
koyil = atten_score[0, 5]
torch.softmax(koyil, -1)
```
